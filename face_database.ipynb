```python
import cv2
from PIL import Image
import numpy as np
import os
```

```python
facedetect=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
cam=cv2.VideoCapture(0)
k=0
id=input("enter user id")
while(True):
    ret,img=cam.read()
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=facedetect.detectMultiScale(gray,1.3,5)
    for(x,y,w,h)in faces:
        #cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
        print(img[y-35:y+h+20,x-10:x+w+10].shape)
        #pil_IMG=Image.fromarray(img[y-35:y+h+20,x-10:x+w+10])
        #IMG = pil_IMG.resize((100,100))
        name=str(id)+"."+str(k)+ ".jpg"
        path=r"C:\Users\user\Documents\anaconda scripts\project 2.0\Face_Database\arjun"
        cv2.imwrite(os.path.join(path,name),img[y-50:y+h+40,x-10:x+w+10])
        #pil_IMG.save(os.path.join(path,name))
    cv2.imshow("Face",img)
    k=k+1
    if k==300:
        break;
```

```python
cam.release()
cv2.destroyAllWindows()
```
